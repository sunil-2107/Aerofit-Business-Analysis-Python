#**Probability  Marginal & Conditional Probability**

**Marginal Probability**

Marginal probability represents the overall probability of purchasing each treadmill, regardless of customer characteristics.

Example:

Probability(Customer buys KP281)  Highest

Probability(Customer buys KP481)  Moderate

Probability(Customer buys KP781)  Lowest

**Insight:**

Most customers prefer entry-level treadmills, indicating a larger beginner market.

 **Conditional Probability**

Conditional probability shows the likelihood of purchasing a product given a specific condition.

Examples:

P(KP781 | Male) is higher than P(KP781 | Female)

P(KP781 | High Income) is significantly high

P(KP281 | Low Fitness) is dominant

P(KP781 | High Fitness) is very high

**Insight:**

Income, gender, and fitness level strongly influence treadmill selection.
